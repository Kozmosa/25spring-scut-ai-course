```python
import seaborn as sns
import matplotlib.pyplot as plt

# Iris
iris = sns.load_dataset("iris")

# 
sns.jointplot(
    data=iris,
    x="petal_length",  # X
    y="petal_width",   # Y
    kind="reg",        # 
    scatter_kws={"s": 20, "alpha": 0.6, "edgecolor": "w"},  # 
    line_kws={"color": "red", "lw": 2},                    # 
    marginal_kws={"bins": 15, "kde": True}                 # 
)

# 
plt.tight_layout()
plt.show()
```
