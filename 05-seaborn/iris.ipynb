```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Iris数据集
iris = sns.load_dataset("iris")

# 创建联合分布图（使用回归分析）
sns.jointplot(
    data=iris,
    x="petal_length",  # X轴特征（花瓣长度）
    y="petal_width",   # Y轴特征（花瓣宽度）
    kind="reg",        # 使用回归分析
    scatter_kws={"s": 20, "alpha": 0.6, "edgecolor": "w"},  # 散点图参数
    line_kws={"color": "red", "lw": 2},                    # 回归线参数
    marginal_kws={"bins": 15, "kde": True}                 # 边缘分布参数
)

# 调整布局并显示图形
plt.tight_layout()
plt.show()
```
