```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  定义横纵坐标数据 (Define X and Y coordinates data)
x = np.array([10, 20, 40, 60, 100])
y = np.array([40, 20, 50, 70, 10])

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['/home/kozumi/fonts/maple_mono.ttf']  # 替换为你的字体文件路径
plt.rcParams['axes.unicode_minus'] = False

# 绘制散点图，颜色设置为蓝色 (Plot scatter plot with blue color)
plt.scatter(x, y, color='blue')

# 添加标题和轴标签 (Add title and axis labels)
plt.title('散点图 (Scatter Plot)')
plt.xlabel('横坐标 (X-axis)')
plt.ylabel('纵坐标 (Y-axis)')

# 显示图形 (Show the plot)
plt.show()
```
